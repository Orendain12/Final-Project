```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to change the path if needed.)
cause_data_to_load = "causes_of_death.csv"
prevalence_data_to_load = "U.S._Chronic_Disease_Indicators_v2.csv"
```
