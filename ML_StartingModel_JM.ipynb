```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('./causes-of-death.csv'))
df.head()
```

```python
plt.bar(df['Year'], df['Age-adjusted Death Rate'])
plt.title('Age-Adjusted Death Rate from 1999 to 2017', fontsize = 14)
plt.xlabel('Year')
plt.ylabel('Age-adjusted Death Rate')
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Year'], df['Deaths'])
plt.title('Deaths from 1999 to 2017', fontsize = 14)
plt.xlabel('Year')
plt.ylabel('Deaths')
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Cause Name'], df['Age-adjusted Death Rate'])
plt.title('Age-Adjusted Death Rate by Cause of Death', fontsize = 14)
plt.xlabel('Cause of Death')
plt.ylabel('Age-adjusted Death Rate')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
df.dtypes
```

```python
all_US = df[df["State"] == "United States"]
all_US.head(5)
```

```python
all_US.dtypes
```

```python
plt.bar(all_US['Cause Name'], all_US['Age-adjusted Death Rate'])
plt.title('Age-Adjusted Death Rate by Cause of Death, entire US', fontsize = 14)
plt.xlabel('Cause of Death')
plt.ylabel('Age-adjusted Death Rate')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.scatter(all_US.Year, all_US.Deaths)
plt.xlabel('Year')
plt.ylabel('Deaths')
plt.show()
```

```python
all_US.dtypes
```

```python
all_US["Age-adjusted Death Rate"] = all_US['Age-adjusted Death Rate'].astype('float64')
all_US.dtypes
```
