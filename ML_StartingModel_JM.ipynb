```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('./causes-of-death.csv'))
df.head()
```

```python
plt.bar(df['Year'], df['Age-adjusted Death Rate'])
plt.title('Age-Adjusted Death Rate from 1999 to 2017', fontsize = 14)
plt.xlabel('Year')
plt.ylabel('Age-adjusted Death Rate')
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Year'], df['Deaths'])
plt.title('Deaths from 1999 to 2017', fontsize = 14)
plt.xlabel('Year')
plt.ylabel('Deaths')
plt.grid(True)
plt.show()
```
